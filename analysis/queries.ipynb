# Queries — Case Técnico Data Architect iFood

Este notebook contém as respostas para as duas queries do case.
Os dados são lidos diretamente da camada Gold via AWS Athena.

**Database:** `ifood_case_gold`  
**Queries SQL:** `analysis/queries.sql`

## 1. Importações e configurações

```python
import logging
import warnings
import awswrangler as wr
import pandas as pd

warnings.filterwarnings("ignore", category=UserWarning, module="awswrangler")

logging.basicConfig(
    level=logging.INFO,
    format="%(asctime)s [%(levelname)s] queries - %(message)s",
    datefmt="%d-%m-%Y %H:%M:%S",
)
logger = logging.getLogger("queries")

logging.getLogger("boto3").setLevel(logging.WARNING)
logging.getLogger("botocore").setLevel(logging.WARNING)
logging.getLogger("awswrangler").setLevel(logging.WARNING)
logging.getLogger("urllib3").setLevel(logging.WARNING)

pd.set_option('display.max_columns', None)
pd.set_option('display.float_format', '{:.2f}'.format)

DATABASE  = "ifood_case_gold"
S3_OUTPUT = "s3://ifood-case-data-lake-kaique/athena-results/"

logger.info("Bibliotecas carregadas com sucesso")
```

## Query 1
**Qual a média de valor total (total_amount) recebido em um mês considerando todos os yellow táxis da frota?**

```python
QUERY_1 = """
    SELECT
        mes,
        avg_total_amount
    FROM ifood_case_gold.table_avg_total_amount_gold
    ORDER BY mes
"""

logger.info("Executando Query 1")

df_query1 = wr.athena.read_sql_query(
    sql=QUERY_1,
    database=DATABASE,
    s3_output=S3_OUTPUT,
)

logger.info("Query 1 concluida | meses=%s", len(df_query1))

display(
    df_query1.style
    .set_caption("Query 1 — Media de total_amount por Mes (Yellow Taxi)")
    .format("{:.2f}", subset=["avg_total_amount"])
    .bar(subset=["avg_total_amount"], color="#cfe2ff", vmin=0)
    .set_properties(**{"text-align": "right"})
)
```

## Query 2
**Qual a média de passageiros (passenger_count) por cada hora do dia que pegaram táxi no mês de maio considerando todos os táxis da frota?**

```python
QUERY_2 = """
    SELECT
        hora,
        avg_passenger_count
    FROM ifood_case_gold.table_avg_passengers_gold
    ORDER BY hora
"""

logger.info("Executando Query 2")

df_query2 = wr.athena.read_sql_query(
    sql=QUERY_2,
    database=DATABASE,
    s3_output=S3_OUTPUT,
)

logger.info("Query 2 concluida | horas=%s", len(df_query2))

display(
    df_query2.style
    .set_caption("Query 2 — Media de passenger_count por Hora em Maio (Todos os Taxis)")
    .format("{:.2f}", subset=["avg_passenger_count"])
    .bar(subset=["avg_passenger_count"], color="#d1e7dd", vmin=0)
    .set_properties(**{"text-align": "right"})
)
```
